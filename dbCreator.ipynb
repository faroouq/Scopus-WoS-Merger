# DB Creator

# Introduction
This Python code depends on the <i><b>"ScopusWoSComparator.ipynb"</i></b>. It takes the file <i><b>"papersToRead.txt"</i></b> generated by the aforementioned notebook and the other two co-located files <i><b>"scopus_title_keyword_abstract_tab.csv"</i></b> and <i><b>"wos_full_record.csv"</i></b>, which were search results exported from Scopus and Web of Science (WoS), respectively.

Db Creator (<i><b>"dbCreator.ipynb"</i></b>) merges the two document and develop a Scopus-like csv file that can be used by <a href="https://www.vosviewer.com/">VOSviewer</a>. The resulting file is created in the smae folder with the jupyter notebook and it is named <i><b>"idsPapersDataset.csv"</i></b>.

## Code

```python
toRead = open("papersToRead.txt", "r")
toReadList = toRead.readlines()
toRead.close()
print("Papers to Read = ", len(toReadList))

wosCount=0
wosLine=[]       # Title, Abstract, Author Keyword, Keyword Plus
wosRecords=[]
wosFile = open("./wos_full_record.csv", "r")
record=wosFile.readline()                               #removing header
record=wosFile.readline()

while(record):
    splittedRecord=record.split('\t')
    wosLine.append(splittedRecord[8].lower().strip())   #title
    wosLine.append(splittedRecord[21].lower().strip())  #Abstract
    wosLine.append(splittedRecord[19].lower().strip())  #Author keyword
    wosLine.append(splittedRecord[20].lower().strip())  #Keyword Plus
    wosRecords.append(wosLine)                          #record it
    wosCount+=1
    wosLine=[]                                          #ready for next loop
    record=wosFile.readline()

print("Web of  Science Count = %d (%d)"% (wosCount, len(wosRecords)))
wosFile.close()


scopusFile= open("./scopus_title_keyword_abstract_tab.csv", "r")
scopusCount=0
scopusLine=[]
scopusRecords=[]

record=scopusFile.readline()                              #removing header
record=scopusFile.readline()

while(record):
    splittedRecord=record.split('\t')
    scopusLine.append(splittedRecord[0].lower().strip())  #title
    scopusLine.append(splittedRecord[2].lower().strip())  #Abstract
    scopusLine.append(splittedRecord[3].lower().strip())  #Author keyword
    scopusLine.append(splittedRecord[4].lower().strip())  #Keyword Plus
    scopusRecords.append(scopusLine)                      #record it
    scopusCount+=1
    scopusLine=[]                                          #ready for next loop
    record=scopusFile.readline()
    
print("Scopus Count = %d (%d)"% (scopusCount, len(scopusRecords)))
scopusFile.close()


found=0                                                  #flag for when the title is found
wosFind=0
scopusFind=0
tempFound=0
theDb=[]
tempDb=[]
temp=[]
cnt=0
for i in range(0, len(toReadList)):
    for j in range(0, wosCount):                                #check for the title      
        if (toReadList[i].strip() == wosRecords[j][0].strip()): #get the sentence
            if (wosRecords[j][1].strip()!=""):
                found = 1
                tempFound=0
                theDb.append(wosRecords[j])
                wosFind+=1
                break
            else:
                print("Empty Abstract")
                tempFound = 1
                found=0
                tempDb=wosRecords[j]                            #Just in case we can't find it scopus
                break
                
            
    if (found==0):
        for k in range(0, scopusCount):
            if (toReadList[i].strip() == scopusRecords[k][0].strip()):
                scopusFind+=1
                found=1
                tempFound=0
                theDb.append(scopusRecords[k])
                break 


                
                
    if (found==0 and tempFound==1):
        theDb.append(tempDb)
        print("Append anyway\n",tempDb)
        
        
    found=0
    tempFound=0
                
print("Combined database size = %d"% (len(theDb)))  
print("WoS Find: ", wosFind)
print("Scopus Find", scopusFind)

titleCount=0
abstractCount=0
authKeyword=0
keyword=0
theDataset=open("./idsPapersDataset.csv", "w")
theDataset.write("Title\tLink\tAbstract\tAuthor Keywords\tIndex Keywords\n")
for i in range(0, len(theDb)):  # Title, Abstract, Author Keyword, Keyword Plus
    if(theDb[i][0]==""):
        titleCount+=1
    if(theDb[i][1]==""):
        abstractCount+=1
    if(theDb[i][2]==""):
        authKeyword+=1
    if(theDb[i][2]==""):
        keyword+=1
        
    theDataset.write('%s\t \t%s\t%s\t%s\n'%(theDb[i][0],theDb[i][1],theDb[i][2],theDb[i][3]))
        
theDataset.close()    
```
